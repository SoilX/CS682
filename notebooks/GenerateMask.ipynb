```python
# k is camera instrinsic matrix
k = np.array([[2304.5479, 0,  1686.2379],
           [0, 2305.8757, 1354.9849],
           [0, 0, 1]], dtype=np.float32)
```

```python
# convert euler angle to rotation matrix
def euler_to_Rot(yaw, pitch, roll):
    Y = np.array([[cos(yaw), 0, sin(yaw)],
                  [0, 1, 0],
                  [-sin(yaw), 0, cos(yaw)]])
    P = np.array([[1, 0, 0],
                  [0, cos(pitch), -sin(pitch)],
                  [0, sin(pitch), cos(pitch)]])
    R = np.array([[cos(roll), -sin(roll), 0],
                  [sin(roll), cos(roll), 0],
                  [0, 0, 1]])
    return np.dot(Y, np.dot(P, R))
```

```python
def draw_obj(image, vertices, triangles):
    for t in triangles:
        coord = np.array([vertices[t[0]][:2], vertices[t[1]][:2], vertices[t[2]][:2]], dtype=np.int32)
        cv2.fillConvexPoly(image, coord, (0,0,255))
       # cv2.polylines(image, np.int32([coord]), 1, (0,0,255))
```

```python
# Load a 3D model of a car
with open('../Code/pku-autonomous-driving/car_models_json/dazhong.json') as json_file:
    data = json.load(json_file)
vertices = np.array(data['vertices'])
vertices[:, 1] = -vertices[:, 1]
triangles = np.array(data['faces']) - 1
```

```python
overlay = np.zeros_like(img)
#count = 1 # this is the first model
for yaw, pitch, roll, x, y, z in zip(yaws, pitches, rolls, xs, ys, zs):
    yaw, pitch, roll, x, y, z = [float(x) for x in [yaw, pitch, roll, x, y, z]]
    # I think the pitch and yaw should be exchanged
    yaw, pitch, roll = -pitch, -yaw, -roll
    Rt = np.eye(4)
    t = np.array([x, y, z])
    Rt[:3, 3] = t
    Rt[:3, :3] = euler_to_Rot(yaw, pitch, roll).T
    Rt = Rt[:3, :]
    P = np.ones((vertices.shape[0],vertices.shape[1]+1))
    P[:, :-1] = vertices
    P = P.T
    img_cor_points = np.dot(k, np.dot(Rt, P))
    img_cor_points = img_cor_points.T
    img_cor_points[:, 0] /= img_cor_points[:, 2]
    img_cor_points[:, 1] /= img_cor_points[:, 2]
    draw_obj(overlay, img_cor_points, triangles)

plt.figure(figsize=(20,20))
plt.imshow(overlay)

alpha = .5
img = np.array(img)
#plt.imshow(img);
cv2.addWeighted(overlay, alpha, img, 1 - alpha, 0, img)
plt.figure(figsize=(20,20))
plt.imshow(img);
print(img.shape)
```

```python
#this are the functions for processing all the images
def DataGen(model_types, yaws, pitches, rolls, xs, ys, zs):
    """This function generates the data for generating mask"""
    count = 0
    for model_type, yaw, pitch, roll, x, y, z in zip(model_types, yaws, pitches, rolls, xs, ys, zs):
        model_type, yaw, pitch, roll, x, y, z = [float(x) for x in [model_type, yaw, pitch, roll, x, y, z]]
        #distance of the car to the camera
        distance = np.sqrt(x**2+y**2+z**2)
        if count == 0:
            data_position = np.array([[model_type, yaw, pitch, roll, x, y, z, distance]])
            count = 1
        else: 
            temp = np.array([[model_type, yaw, pitch, roll, x, y, z, distance]])
            data_position = np.concatenate((data_position, temp), axis = 0)
        
    # sort the data based on distance in a descending order
    data_position = data_position[(-data_position[:,7]).argsort()]
    #print(data_position)
    return data_position

def MaskGen(img, data_position):
    """this function generate the mask based on the img and the data
    mask the car far away first, then mask the car nearby"""
    overlay = np.zeros_like(img)
    for i in range(len(data_position)):
        #load the 3D model
        model_num = int(data_position[i, 0])
        a = car_id2name[model_num].name
        path = '../Code/pku-autonomous-driving/car_models_json/' + a + '.json'
        with open(path) as json_file:
            data = json.load(json_file)
        vertices = np.array(data['vertices'])
        vertices[:, 1] = -vertices[:, 1]
        triangles = np.array(data['faces']) - 1
        
        # do the transformation from 3D to 2D projection
        yaw, pitch, roll, x, y, z = data_position[i, 1], data_position[i, 2], data_position[i, 3], data_position[i, 4], data_position[i, 5], data_position[i, 6]
        temp = np.zeros_like(img)
        # pitch and yaw should be exchanged
        yaw, pitch, roll = -pitch, -yaw, -roll
        Rt = np.eye(4)
        t = np.array([x, y, z])
        Rt[:3, 3] = t
        Rt[:3, :3] = euler_to_Rot(yaw, pitch, roll).T
        Rt = Rt[:3, :]
        P = np.ones((vertices.shape[0],vertices.shape[1]+1))
        P[:, :-1] = vertices
        P = P.T
        img_cor_points = np.dot(k, np.dot(Rt, P))
        img_cor_points = img_cor_points.T
        img_cor_points[:, 0] /= img_cor_points[:, 2]
        img_cor_points[:, 1] /= img_cor_points[:, 2]
        draw_obj(temp, img_cor_points, triangles)
        # change it to boolean mask
        mask = (temp == 255)
        #print(mask.shape)
        overlay[mask] = model_num
    
    return overlay[:,:,2]

# data_position = DataGen(model_types, yaws, pitches, rolls, xs, ys, zs)
# segmented = MaskGen(img, data_position)
# plt.figure(figsize=(20,20))
# plt.imshow(segmented*5)
```
